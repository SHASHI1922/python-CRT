```python
#DIGITAL TIME
import time
import sys
h=3
s=0
m=40
while True:
  sys.stdout.write("\r{:02d}:{:02d}:{:02d}".format(h,m,s))
  sys.stdout.flush()
  time.sleep(1)
  s=s+1
  if s==60:
    s=0
    m=m+1
  if m==60:
    m=0
    h=h+1
  if h==24:
    h=0
```

```python
#
names = ["starck","gruce","wanda","vision","soooop"]
marks = [30,40,50],[60,70,80],[99,88,90],[99,89,56],[45,67,88]
for i in range (5):
  p = sum (marks [i])//3
  if p >80 :
    g = "s"
  elif p<40:
    g = "f"
  elif p>40 and p<60 :
    g = "b"
  else :
      g = "A"
  print ("{}.mr/ms {} has scored {}%-{} Grade ".format(i+1,names [i].title(),p,g))
```

```python
#zip (merging)
names = ["starck","gruce","wanda","vision","soooop"]
age = [24,45,55,65,45]
res = list(zip(names, age ))
res1 = sorted(res)
pos =1
for i in res1:
  print ("Rank {}: {} age is {} yrs old".format(pos,i[1],i[0]))
  pos+=1
```

```python
a=int (input("enter no of taeam "))
teams = []
for i in range (a):
  t = input("enter team name : ")
  teams.append(t)
m=int (input("enter no of taeam "))

matches =[]
pos =1
for i in range (0,a-1):
  for j in range (i+1,a):
    for k in range (m):
      matches.append([teams[i],teams[j]])
for i in matches:
  print("match {} : {} vs {}".format(pos,i[0],i[1]))
  pos+=1
```
